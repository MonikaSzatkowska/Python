```python
import pandas as pd

baby_names = pd.read_csv('C:\\Users\Admin\Desktop\imiona2000-2019.csv')

print(baby_names.columns)

baby_names.columns = ['Year', 'Name', 'Count', 'Sex']

print(baby_names.head())

print(baby_names.info())
```

```python
girls_names = baby_names[baby_names['Sex']=='K'].reset_index(drop=True)

girls_names.head()
```

```python
boys_names = baby_names[baby_names['Sex']=='M'].reset_index(drop=True)

boys_names.head()
```

```python
least_popular_g_names = girls_names.drop(columns='Year').groupby('Name').sum().sort_values(by='Count').reset_index()

least_popular_g_names = least_popular_g_names[least_popular_g_names.Count < 5]

least_popular_g_names_alphabetical = {}

for name in set(least_popular_g_names.Name):
    if name[:1] in least_popular_g_names_alphabetical.keys():
        least_popular_g_names_alphabetical[name[:1]].append(name)
    else: least_popular_g_names_alphabetical[name[:1]] = [name]
        
print(len(least_popular_g_names))
print(sorted(least_popular_g_names_alphabetical.items()))
```

```python
least_popular_b_names = boys_names.drop(columns='Year').groupby('Name').sum().sort_values(by='Count').reset_index()

least_popular_b_names = least_popular_b_names[least_popular_b_names.Count < 5]

least_popular_b_names_alphabetical = {}

for name in set(least_popular_b_names.Name):
    if name[:1] in least_popular_b_names_alphabetical.keys():
        least_popular_b_names_alphabetical[name[:1]].append(name)
    else: least_popular_b_names_alphabetical[name[:1]] = [name]
        
print(len(least_popular_b_names))
print(sorted(least_popular_b_names_alphabetical.items()))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

years = girls_names.Year.unique()

unique_names_count_g = {}

for year in years:
    unique_names_count_g[year] = len(girls_names.Name[girls_names.Year == year].unique())
    
unique_names_count_g = pd.DataFrame(unique_names_count_g.items(), columns=['Year', 'Names Count'])

unique_names_count_g.Year = pd.to_datetime(unique_names_count_g.Year, format='%Y')

unique_names_count_g.set_index('Year', inplace=True)

unique_names_count_g.plot()

plt.show()

girls_names_2012 = set(girls_names.Name[girls_names.Year == 2012])
girls_names_2013 = set(girls_names.Name[girls_names.Year == 2013])

new_names_g = girls_names_2013.difference(girls_names_2012)

print(len(new_names_g), new_names_g)
```

```python
girls_names_top_50_year = {}

girls_names_top_10_year = {}

girls_names.sort_values(by='Count', ascending=False)

for year in years:
    names_50 = girls_names[['Name','Count']][girls_names.Year == year][:50]
    names_10 = girls_names[['Name','Count']][girls_names.Year == year][:10]
    girls_names_top_50_year[year] = names_50.reset_index()
    girls_names_top_10_year[year] = names_10.reset_index()
```

```python
freq_50 = {}
freq_10 = {}

for year in girls_names_top_50_year.values():
    
    names = list(year.Name)
    
    for name in names:  
        
        if name in freq_50:
            freq_50[name] += 1
        else: freq_50[name] = 1

freq_50_sorted = pd.DataFrame(sorted(freq_50.values())).hist(bins = 20)

for year in girls_names_top_10_year.values():
    
    names = list(year.Name)
    
    for name in names:  
        
        if name in freq_10:
            freq_10[name] += 1
        else: freq_10[name] = 1

top_10_names = []

for name, count in freq_10.items():
        if count >= 15:
            top_10_names.append((count, name))
            
print(len(top_10_names), top_10_names)
      
top_names = []

for name, count in freq_50.items():
    
    if count == 20:
        top_names.append(name)
        
top_names.sort()

print(len(top_names),top_names)
```

```python
total_50 = {}

for year, dic in girls_names_top_50_year.items():
    
    names = (dic.Name)
    count = list(dic.Count)
    
    for i, name in enumerate(names):  
        if name in total_50:
            total_50[name] += count[i]
        else: total_50[name] = count[i]

print(total_50.items())

total_50_count = pd.DataFrame(total_50.items(), columns=['Name','Sum'])

total_50_count.plot.box()
plt.show()

total_50_count.sort_values(by='Sum', ascending=False)[:10].plot.bar(x='Name')
```

```python
import numpy as np

fig, axes = plt.subplots(5, 4, figsize=(25,25))

plt.subplots_adjust(hspace=0.4)

shape = (5,4)
years_keys = np.reshape(years,shape)

for i in range(0,5):
    for j in range(0,4):
        df = girls_names_top_50_year[years_keys[i][j]]
        axes[i,j].bar(df.iloc[:10,1], df.iloc[:10,2])
        axes[i,j].set_title('Rok ' + str(years_keys[i][j]))
        for tick in axes[i,j].get_xticklabels():
            tick.set_rotation(45)

plt.show()
```
